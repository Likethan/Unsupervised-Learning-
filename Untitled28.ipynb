```python
#  Step 1: Import Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans, DBSCAN
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

#  Step 2: Load Dataset
iris = load_iris()
X = pd.DataFrame(iris.data, columns=iris.feature_names)
y_true = iris.target

#  Step 3: Feature Scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

#  Step 4: Dimensionality Reduction (for visualization)
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# --- K-MEANS CLUSTERING ---
#  Step 5a: Apply K-Means
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans_labels = kmeans.fit_predict(X_scaled)

#  Step 6a: Visualize K-Means Clusters
plt.figure(figsize=(8,4))
plt.subplot(1,2,1)
sns.scatterplot(x=X_pca[:,0], y=X_pca[:,1], hue=kmeans_labels, palette='viridis')
plt.title("K-Means Clustering")
plt.xlabel("PCA1")
plt.ylabel("PCA2")

# --- DBSCAN CLUSTERING ---
#  Step 5b: Apply DBSCAN
dbscan = DBSCAN(eps=0.6, min_samples=5)
dbscan_labels = dbscan.fit_predict(X_scaled)

#  Step 6b: Visualize DBSCAN Clusters
plt.subplot(1,2,2)
sns.scatterplot(x=X_pca[:,0], y=X_pca[:,1], hue=dbscan_labels, palette='tab10')
plt.title("DBSCAN Clustering")
plt.xlabel("PCA1")
plt.ylabel("PCA2")

plt.tight_layout()
plt.show()
```
